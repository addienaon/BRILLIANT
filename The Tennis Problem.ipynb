Binomial Distribution - A probability distribution that summarises the likelihood that a variable will take one of two independent values under a given set of parameters. Obtained via Bernoulli trials, assuming these criteria are met:
- There must only be 2 possible outcomes (success ad failure)
- Each outcome has a fixed probability of occuring. A success has the probability of p, and a failure has the probability of (1-p).
- Each trial is completely independent of all others. 

```python
from scipy.stats import binom
# setting the values
# of n and p
n = 4
p = 0.6
# defining the list of r values
r_values = list(range(n + 1))
# obtaining the mean and variance 
mean, var = binom.stats(n, p)
# list of pmf values
dist = [binom.pmf(r, n, p) for r in r_values ]
# printing the table
print("r\tp(r)")
for i in range(n + 1):
    print(str(r_values[i]) + "\t" + str(dist[i]))
# printing mean and variance
print("mean = "+str(mean))
print("variance = "+str(var))
```

```python
import matplotlib.pyplot as plt
plt.bar(r_values, dist)
plt.xlabel('# of success (r)')
plt.ylabel('Prob. of acheving r success and n-r failure (PMF)') #Probability mass failure
plt.show()
```

Pascals Triangle - A triangular array of type binomial coefficients that arises in probability theory, cobinatorics, and algebra.
Binomial coefficient - an indexed pair of integers n>=k>=0 and is written:

                                                                            (n)
                                                                            (k) 
                                                                            
It is usually read "n choose k" because there are (n k) ways to choose an unordered subset of k elements from a fixed set of n elements. 
https://en.wikipedia.org/wiki/Pascal's_triangle?msclkid=e3fa55feb68511ecb5b089e3cd4db5bd

https://en.wikipedia.org/wiki/Binomial_coefficient

```python
# Pascal's Triangle (n = number of rounds. Using 4-1 example)
# This constructs the triangle, then you have to plug in the n and k values to get the number of combinations (which is stored as a tuple in each triangle cell!)
binom = {}
for n in range(0, 10):
    for k in range(0, n + 1):
        if k == 0 or k == n:
            binom[(n, k)] = 1
        else:
            binom[(n, k)] = binom[(n - 1, k)] + binom[(n - 1, k - 1)]
            
combinations = binom[4,1] #4 wins, 1 loss
print(combinations) #4-combinations!
```

```python
# Compute Win Probability
#x is k in the Tennis problem
#

def prob(n, p):
    total = 0
    for x in range(0, n):
        total += ((1 - p) ** x) * binom[(n - 1 + x, x)] 
    total *= (p ** n)
    return total
```
